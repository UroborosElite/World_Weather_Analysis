```python
# Starting URL for Weather Map API Call.
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
#print(url)
# Create an endpoint URL for a city.
city_url = url + "&q=" + "boston"
print(city_url)
```

```python
# Make a 'Get' request for the city weather.
city_weather = requests.get(city_url)
city_weather
# Get the JSON text of the 'Get' request.
city_weather.json()
```

```python
# Get the JSON data.
boston_data = city_weather.json()
```

```python
boston_data['sys']["country"]
```

```python
lat = boston_data["coord"]["lat"]
lng = boston_data["coord"]["lon"]
max_temp = boston_data["main"]["temp_max"]
humidity = boston_data["main"]["humidity"]
clouds = boston_data["clouds"]["all"]
wind = boston_data["wind"]["speed"]
description = boston_data["weather"][0]["description"]

print(lat, lng, max_temp, humidity, clouds, wind, description)
```

```python
boston_data['dt']
```

```python
city_data_all = []
for i, city in enumerate(cities):
    if i % 50 == 0:
        print("sleeping for open weather data")
        time.sleep(60) 
    try:
        city_url = url + "&q=" + city
        #print(city_url)
        city_weather = requests.get(city_url)
        if city_weather.status_code != 200:
            raise Exception("city not found")
        city_data = city_weather.json()
        city_country = city_data["sys"]["country"]
        at = city_data["coord"]["lat"]
        lng = city_data["coord"]["lon"]
        max_temp = city_data["main"]["temp_max"]
        humidity = city_data["main"]["humidity"]
        clouds = city_data["clouds"]["all"]
        wind = city_data["wind"]["speed"]
        description = city_data["weather"][0]["description"]    
        # Convert the date to ISO standard.
        city_date = datetime.utcfromtimestamp(city_data["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        # Append the city information into city_data list.
        city_data_all.append({"City": city,
                            "Country": city_country,
                            "Lat": at,
                            "Lng": lng,
                            "Max Temp": max_temp,
                            "Humidity": humidity,
                            "Cloudiness": clouds,
                            "Wind Speed": wind,
                            "Current Description": description
                             })
                            ##"Date": city_date})

    # If an error is experienced, skip the city.   
    except Exception as e: 
        print(e,i,city_url,city)
        print("City not found. Skipping...")

# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
print(city_data_all)
```

```python
# Convert the array of dictionaries to a Pandas DataFrame.
city_data_df = pd.DataFrame(city_data_all)
city_data_df.head(10)
```

```python
# Create the output file (CSV).
output_data_file = "weather_data/WeatherPy_Database.csv"
# Export the City_Data into a CSV.
city_data_df.to_csv(output_data_file, index_label="City_ID")
```
